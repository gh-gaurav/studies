```python
import pandas as pd
```

```python
gods_data = {
    'Name': ['Zeus', 'Poseidon', 'Hades', 'Apollo', 'Ares', 'Hephaestus', 'Hermes', 'Dionysus', 'Heracles', 'Pan'],
    'Age': [10000, 9000, 8500, 7500, 8000, 8200, 8800, 7600, 9200, 7800],
    'Domain': ['Sky', 'Sea', 'Sun', 'Sun', 'War', 'Sky', 'Love', 'Sea', 'War', 'Sun'],
    'Symbol': ['Thunderbolt', 'Trident', 'Helmet of Invisibility', 'Lyre', 'Sword', 'Hammer', 'Caduceus', 'Grapevine', 'Club', 'Pan flute'],
    'Abode': ['Mount Olympus', 'Undersea Palace', 'Underworld', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus']
}
```

```python
goddesses_data = {
    'Name': ['Athena', 'Hera', 'Aphrodite', 'Artemis', 'Demeter', 'Persephone', 'Hestia', 'Hecate', 'Nemesis', 'Eos'],
    'Age': [8500, 9500, 8200, 7800, 8300, 8400, 8700, 9100, 9300, 9400],
    'Domain': ['Wisdom', 'Sun', 'Love', 'Hunting', 'Sky', 'Love', 'Sun', 'Magic', 'War', 'Dawn'],
    'Symbol': ['Owl', 'Peacock', 'Dove', 'Bow and arrow', 'Wheat', 'Pomegranate', 'Fire', 'Torch', 'Sword', 'Rosy-fingered'],
    'Abode': ['Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Mount Olympus', 'Underworld', 'Mount Olympus', 'Underworld', 'Mount Olympus', 'Sky']
}
```

```python
god_df = pd.DataFrame(gods_data)
```

```python
goddess_df = pd.DataFrame(goddesses_data)
```

```python
goddess_df
```

```python
#task1(create a new table that includes information about both gods and goddesses.)
```

```python
both_gods = pd.merge(god_df,goddess_df,how = 'outer')
```

```python
both_gods
```

```python
#task_2(Filter the merged table to only include gods and goddesses who are older than 8000 years, then sort them based on their ages in descending order.)
```

```python
task_2 = both_gods[both_gods['Age']>8000].sort_values(by = 'Age',ascending = False).reset_index(drop = True)
```

```python
task_2
```

```python
#task_3(Join the two tables based on the "Domain" field and calculate the average age of gods and goddesses in each domain)
```

```python
task_3 = both_gods.groupby('Domain')['Age'].mean()
```

```python
task_3
```

```python
#task_4(Determine which god/goddess has the highest age, and then find out if they are a god or a goddess.)
```

```python
max_age = both_gods.loc[both_gods['Age'].idxmax()]
```

```python
if max_age['Name'] in both_gods['Name'].values:
    print("it is the god")
else:
    print("it is the goddess")
```

```python
#task_5(Create a new column in each table called "Age Group" and categorize the gods/goddesses into groups such as "Young" (age< 5000), "Middle-aged" (age between 5000 and 8000), and "Old" (age > 8000).)
```

```python
bins = [0,5000,8000,float('inf')]
labels = ['Young','Middle-aged','Old']
god_df['Age Group'] = pd.cut(god_df['Age'],bins = bins,labels = labels,right = 'False')
goddess_df['Age Group'] = pd.cut(goddess_df['Age'],bins = bins,labels = labels,right = 'False')
```

```python
god_df
```

```python
goddess_df
```

```python
#task_6(Compare the average ages of gods and goddesses. Is there a significant age difference between them? If yes, which group tends to be older?)
```

```python
avg_god = god_df['Age'].mean()
avg_goddess = goddess_df['Age'].mean()
```

```python
if avg_god > avg_goddess:
    print("gods are tend to be older")
elif avg_goddess > avg_god:
    print("goddess are tend to be older")
else:
    print("there is no difference between the avg age")
```

```python
#task_7(Write a Python program using for loop to iterate over the "Age" column of the merged table (after merging the gods and goddesses tables) and print out the names of gods/goddesses who are older than 8000 years.)
```

```python
for index,row in both_gods.iterrows():
    if row['Age']>8000:
        print(row['Name'])
```

```python
#task_8(Write a Python program to find the oldest god/goddess from the merged table (after merging the gods and goddesses tables) by iterating through the "Age" column using a while loop. Print out the name of the oldest god/goddess and their age.)
```

```python
max_age = 0
name = None
index = 0
while index < len(both_gods):
    age = both_gods.loc[index,'Age']
    if age > max_age:
        max_age = age
        name = both_gods.loc[index,'Name']
    index += 1
print('god name: ',name)
print('age: ',max_age)
```
